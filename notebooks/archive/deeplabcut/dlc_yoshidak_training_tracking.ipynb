```python
# import os
# os.environ["CUDA_VISIBLE_DEVICES"] = "-1"

from pathlib import Path

import deeplabcut
```

```python
PROJECT_ROOT = Path.cwd().resolve()
project_name = 'FishPoseYOSHIDAK-Round1'
experimenter = 'Lucas'
working_directory = PROJECT_ROOT / 'DeepLabCut'
config_path = str(working_directory / f'{project_name}-{experimenter}' / 'config.yaml')
```

```python

deeplabcut.extract_frames(config_path, mode='automatic', algo='kmeans', userfeedback=False, crop=False)
```

```python

# deeplabcut.check_labels(config_path)

deeplabcut.create_training_dataset(config_path)

# Adjust 'shuffle=1' or 'displayiters', 'saveiters' as needed
deeplabcut.train_network(config_path, shuffle=1, displayiters=500, saveiters=10000, maxiters=50000)

deeplabcut.evaluate_network(config_path, plotting=True)

```

```python
behavior_root = PROJECT_ROOT / 'data/raw/yoshidak/behavior'
videos = [str(p) for p in sorted(behavior_root.rglob('*.avi'))]
videos = [v for v in videos if '-before' not in v and '-after' not in v]

deeplabcut.analyze_videos(config_path, videos, save_as_csv=True)
# deeplabcut.create_labeled_video(config_path, videos)
```
